```python
with open('claseDump.json','r') as f:
    clases=json.loads(f.read())
```

```python
clases=g.clases
```

```python
# Usamos solo el nombre y clave de cada clase
aSubir={}
for claveClase, datosClase in clases.items():
    depto=claveToDepto[claveClase.split('-')[0]]
    aSubir[datosClase['nombre']]={'depto':depto,'color':4285046584,'asesoriasUids':[]}
aSubir
# aSubir es un diccionario con los datos de las clases a subir al sistema
```

```python
# Subimos datos
cred = credentials.Certificate("key.json")
firebase_admin.initialize_app(cred)
# Inicialización de la app.
# Se requiere verificar las credenciales guardadas en el json con claves.
db=firestore.client()
```

```python
#Upload clases
ref = db.collection(u'clases')
for k,v in aSubir.items():
    ref.document(k).set(v)
```
